```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
# 1. a)
N = 1000
trend = np.array([1/1489*x**2 + 1/1000*x + 1355 for x in range(N)])
sezon = np.array([54*np.sin(1/12*np.pi*x) + 46*np.sin(1/24*np.pi*x) for x in range(N)])
variatii = np.random.normal(0, 35, (N,))

serie = np.array(list(map(lambda x, y, z: x+y+z, trend, sezon, variatii)))

plt.plot(np.arange(N), serie)
plt.title('serie')
plt.show()

plt.plot(np.arange(N), trend)
plt.title('trend')
plt.show()

plt.plot(np.arange(N), sezon)
plt.title('sezon')
plt.show()

plt.plot(np.arange(N), variatii)
plt.title('variatii')
plt.show()
```

```python
# b)
autocorelatie = np.correlate(serie, serie, mode="full")
print(autocorelatie)

plt.plot(np.arange(-autocorelatie.size//2, autocorelatie.size//2), autocorelatie)
plt.show()
```
