```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cerinta 1
gausiana_1d = np.random.normal(521, 32, 1000)
plt.hist(gausiana_1d)
plt.show()
```

```python
gausiana_2d = np.random.multivariate_normal(
    np.array([521, 127]),
    np.array([
        [1, 3/5],
        [3/5, 2]
    ]),
    1000)
plt.scatter(gausiana_2d[:, 0], gausiana_2d[:, 1])
plt.show()
```
