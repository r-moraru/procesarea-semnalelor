```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def getFourierMatrix(N):
    return [
        np.array([math.e**(2j*math.pi*k*m/N) for k in range(N)])
        for m in range(N)
    ]
```

```python
F = getFourierMatrix(8)
fig, axs = plt.subplots(8, figsize=(4,32))

getReal = lambda x : x.real
getImag = lambda x : x.imag
for i in range(8):
    real = getReal(F[i])
    print(real)
    imag = getImag(F[i])
    print(imag)
    print()

    axs[i].plot(np.arange(0, 8, 1), real, label="real")
    axs[i].plot(np.arange(0, 8, 1), imag, label="imag")

plt.show()
```

```python
np.all(np.matmul(F, np.transpose(np.conjugate(F))) - 8*np.identity(8) < 2e-14*(1+1j))
```
