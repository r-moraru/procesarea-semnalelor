```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
def sinA(t):
    return 2 * np.sin(2*np.pi*440*t + np.pi/2)

def cosA(t):
    return 2 * np.cos(2*np.pi*440*t)
```

```python
# ex 1
fig, axs = plt.subplots(2)
plt.suptitle('Comparatie sinus si cosinus')

discrete_time = np.linspace(0.0, pow(2, -8), 100)
axs[0].plot(discrete_time, sinA(discrete_time))
axs[1].plot(discrete_time, cosA(discrete_time))
```

```python
# ex 2 pt 1
def sinC(t, phi):
    return np.sin(2*np.pi*256*t + phi)

def sin0(t):
    return sinC(t, 0)

def sin1(t):
    return sinC(t, np.pi/5)

def sin2(t):
    return sinC(t, np.pi/3)

def sin3(t):
    return sinC(t, np.pi/2)

discrete_time = np.linspace(0.0, pow(2, -7), 50)
fig, axs = plt.subplots(1)
axs.plot(discrete_time, sin0(discrete_time))
axs.plot(discrete_time, sin1(discrete_time))
axs.plot(discrete_time, sin2(discrete_time))
axs.plot(discrete_time, sin3(discrete_time))
```

```python
# ex 2 pt 2

z = np.random.normal(size=50) # Gaussian distribution by default

def calculateGamma(x, z, SNR):
    return np.sqrt(np.linalg.norm(x)**2 / (np.linalg.norm(z)**2 * SNR))

gamma0 = calculateGamma(sin0(discrete_time), z, 0.1)
gamma1 = calculateGamma(sin1(discrete_time), z, 1)
gamma2 = calculateGamma(sin2(discrete_time), z, 10)
gamma3 = calculateGamma(sin3(discrete_time), z, 100)

fig, axs = plt.subplots(1)
axs.plot(discrete_time, sin0(discrete_time) + gamma0 * z)
axs.plot(discrete_time, sin1(discrete_time) + gamma1 * z)
axs.plot(discrete_time, sin2(discrete_time) + gamma2 * z)
axs.plot(discrete_time, sin3(discrete_time) + gamma3 * z)
```
