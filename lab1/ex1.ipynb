```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def x(t):
    # cos(520πt + π/3)
    return np.cos(520*np.pi*t + np.pi/3)

def y(t):
    # cos(280πt − π/3)
    return np.cos(280*np.pi*t - np.pi/3)

def z(t):
    # cos(120πt + π/3)
    return np.cos(120*np.pi*t - np.pi/3)
```

```python
# b)
discrete_time = np.arange(0, 0.03, 0.0005)

fig, axs = plt.subplots(3)
plt.suptitle('comparatie cos')

axs[0].plot(discrete_time, x(discrete_time))
axs[1].plot(discrete_time, y(discrete_time))
axs[2].plot(discrete_time, z(discrete_time))

plt.show()
```

```python
# c)
fs = 200
start_time, end_time = 0, 0.03
discrete_time = np.arange(start_time, end_time, (end_time-start_time)/fs)

fig, axs = plt.subplots(3)
plt.title('Esantionare')

axs[0].plot(discrete_time, x(discrete_time))
axs[1].plot(discrete_time, y(discrete_time))
axs[2].plot(discrete_time, z(discrete_time))

plt.show()
```
