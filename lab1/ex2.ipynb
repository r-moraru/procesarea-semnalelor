```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# a)
def sinusoidal1(t):
    return np.sin(2*np.pi*400*t)

discrete_time = np.linspace(0, 4/400, 1600)

fig, axs = plt.subplots(1)
plt.title('sinosoidal1')

axs.plot(discrete_time, sinusoidal1(discrete_time))

plt.show()
```

```python
# b)

def sinusoidal2(t):
    return np.sin(2*np.pi*800*t)

discrete_time = np.linspace(0, 3, 3*800*100)

fig, axs = plt.subplots(1)
plt.title('sinosoidal2')

axs.plot(discrete_time, sinusoidal2(discrete_time))

plt.show()
```

```python
def sawtooth(t):
    return 2*(240*t - np.floor(240*t)) - 1

discrete_time = np.linspace(0, 4/240, 4*240)

fig, axs = plt.subplots(1)
plt.title('sawtooth')

axs.plot(discrete_time, sawtooth(discrete_time))

plt.show()
```

```python
def square(t):
    return np.sign(np.sin(2*np.pi*300*t))

discrete_time = np.linspace(0, 4/300, 4*300)

fig, axs = plt.subplots(1)
plt.title('sawtooth')

axs.plot(discrete_time, square(discrete_time))

plt.show()
```
