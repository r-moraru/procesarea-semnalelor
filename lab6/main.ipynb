```python
# 2.
def poly_convergence_mul(p_coefs1, p_coefs2):
    return np.polymul(np.poly1d(p_coefs1), np.poly1d(p_coefs2)).coefficients

def poly_convergence_fft(p_coefs1, p_coefs2):
    len_sum = len(p_coefs1) + len(p_coefs2)
    fft1 = np.fft.fft(p_coefs1, len_sum-1)
    fft2 = np.fft.fft(p_coefs2, len_sum-1)
    return np.fft.ifft(fft1 * fft2, len_sum-1)
```

```python
p1 = np.random.randint(-500, 500, size=500)
p2 = np.random.randint(-500, 500, size=500)

fig, axs = plt.subplots(2)
axs[0].plot(np.arange(0, len(p1)+len(p2)-1), poly_convergence_mul(p1, p2))
axs[1].plot(np.arange(0, len(p1)+len(p2)-1), poly_convergence_fft(p1, p2))
```
